### Задача 2 (1 балл)

Преобразуйте дату публикации поста в более привычный формат – добавьте в датафрейм `small` столбец `datetime` с датой типа `datetime`.

**Подсказка.** Код для преобразования одной даты в формате POSIX (число секунд с 1 января 1970 года) в объект `datetime`:

```python
from datetime import datetime

# было: число секунд
in_sec = 1676116200

# стало: объект datetime
dt = datetime.fromtimestamp(in_sec) 
print(dt)
```

**Дополнительно.** Если будете использовать подобную трансформацию в своей работе, ещё один полезный момент – дату в формате `datetime` можно преобразовать в обычную строку по своему шаблону (про обозначения с `%` можно почитать [здесь](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior), в разделе про *Format codes*). Только стоит иметь в виду, что даты в формате `datetime` сортируются хронологически, а строки – по алфавиту, что не всегда удобно.

```python
# обычная строка, отформатированная по шаблону
# Y – четырехзначный год, y - двузначный год
# m – месяц, d - день
# H – часы, M - минуты
# A - день недели, a - день недели сокращенно

in_str = dt.strftime("%Y/%m/%d %H:%M")
print(in_str)

in_str_w = dt.strftime("%Y-%m-%d %A %H:%M")
print(in_str_w)
```
