## Задание

* Глобальная задача: создать небольшое веб-приложение, которое запрашивает у пользователя ключевые слова для поиска книги, находит результаты в каталоге на сайте книжного магазина «Библио-Глобус» и выводит таблицу с названиями и ценами книг, где строки отсортированы по возрастанию или убыванию цены. Для простоты в рамках этого задания будет забирать данные только с первой страницы результатов поиска.

* Поиск и сбор данных осуществляется через `Selenium` и `BeautifulSoup`, приложение создается с помощью библиотеки `streamlit`. Приложение достаточно создать локально на компьютере, необязательно его публиковать на сервере в открытом доступе.

* Результат выполнения домашнего задания – исполняемый файл `bg_app.py` (вспомогательные файлы – при необходимости). 

### Часть 1: Selenium (4 балла)

**(3 балла)** Напишите функцию `get_dataframe(key, sorting)`, которая:

1. Принимает на вход два аргумента, `key` и `sorting`: 

* `key`: строка с названием книги/ключевыми словами для поиска книги;
* `sorting`: параметр сортировки результатов, `True` – по возрастанию цены, `False` – по убыванию цены.
    
2. Возвращает датафрейм с названиями книг и их ценами, полученными на основе первой страницы результатов поиска в каталоге на сайте книжного магазина «Библио-Глобус». Строки в датафрейме должны быть отсортированы по возрастанию или убыванию цены в зависимости от значения `sorting`.

Более подробный алгоритм работы функции:

1. Функция открывает с помощью `Selenium` браузер, переходит на страницу <https://www.biblio-globus.ru>, находит поле для поиска, вводит в него значение из `key`, отправляет запрос.

2. Функция забирает исходный код первой страницы с результатами, извлекает оттуда названия книг и цены, создает на их основе датафрейм. Выполняет сортировку строк по цене, порядок сортировки определяется значением в `sorting`.

**(1 балл)** Потестируйте написанную функцию на каком-нибудь примере, покажите результат исполнения функции.
